```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

#in this model we have increased the no of neighbors 

```python
df=pd.read_csv("citrus.csv")
X1=df.iloc[:,1:3]
y1=df.iloc[:,0]
X_train,X_test,y_train,y_test=train_test_split(X1,y1, random_state=42)
for i in range(1,50):
  Model2=KNeighborsClassifier(n_neighbors=i)
  Model2.fit(X_train,y_train)
  pred1=Model2.predict(X_test)
  print("This is accuracy score of",i,accuracy_score(pred1,y_test))
```

# now here we by the help for loop checking random state

```python
df=pd.read_csv("citrus.csv")
X1=df.iloc[:,1:3]
y1=df.iloc[:,0]
X_train,X_test,y_train,y_test=train_test_split(X1,y1, random_state=i)
for i in range(1,50):
  Model2=KNeighborsClassifier(n_neighbors=5)
  Model2.fit(X_train,y_train)
  pred1=Model2.predict(X_test)
  print("This is accuracy score of",i,accuracy_score(pred1,y_test))
```

```python
#now we doing feature scaling and using min max scaler
```

```python
from sklearn.preprocessing import MinMaxScaler

sc=MinMaxScaler()

X_scale=sc.fit_transform(X1)

X_train2,X_test2,y_train2,y_test2=train_test_split(X_scale,y1, random_state=1)

Model2=KNeighborsClassifier()

Model2.fit(X_train2,y_train2)

pred2=Model2.predict(X_test2)

accuracy_score(pred2,y_test2)
```

```python
from sklearn.datasets import load_wine 
```

```python
wine=load_wine()
```

#now we spilt the dataset into faeture and target 

```python
X3=wine.data
y3=wine.target
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
Model3=KNeighborsClassifier()
X_train3,X_test3,y_train3,y_test3= train_test_split(X3,y3, random_state=42)
Model3.fit(X_train3,y_train3)
# calculating training predition
X_train_pred3=Model3.predict(X_train3)
# calculating testing predition
X_test_pred3=Model3.predict(X_test3)
print("This is training accuracy :", accuracy_score(X_train_pred3, y_train3))
print("This is testing accuracy :", accuracy_score(X_test_pred3, y_test3))
```

```python
#Parameter tunning 
for i in range(2,10):
  print(f"--------------------{i}----------------")
  Model3=KNeighborsClassifier(n_neighbors=i)        
  X_train3,X_test3,y_train3,y_test3= train_test_split(X3,y3, random_state=42)
  Model3.fit(X_train3,y_train3)
  # calculating training predition
  X_train_pred3=Model3.predict(X_train3)
  # calculating testing predition
  X_test_pred3=Model3.predict(X_test3)
  print("This is training accuracy :", accuracy_score(X_train_pred3, y_train3))
  print("This is testing accuracy :", accuracy_score(X_test_pred3, y_test3))
```

```python
#feature scaling 
from sklearn.preprocessing import MinMaxScaler

X4=wine.data
y4=wine.target
sc=MinMaxScaler()
X_scale4=sc.fit_transform(X4)
for i in range(2,20):
  print(f"--------------------------training :{i}--------------------")
  Model4=KNeighborsClassifier(n_neighbors=i)
  X_train4,X_test4,y_train4,y_test4= train_test_split(X_scale4,y4, random_state=42)
  Model4.fit(X_train4,y_train4)
  # calculating training predition
  X_train_pred4=Model4.predict(X_train4)
  # calculating testing predition
  X_test_pred4=Model4.predict(X_test4)
  print("This is training accuracy :", accuracy_score(X_train_pred4, y_train4))
  print("This is testing accuracy :", accuracy_score(X_test_pred4, y_test4))
```

```python
#feature scaling 
#final model to deploy
from sklearn.preprocessing import MinMaxScaler

X4=wine.data
y4=wine.target
sc=MinMaxScaler()
X_scale4=sc.fit_transform(X4)

Model4=KNeighborsClassifier(n_neighbors=6)
X_train4,X_test4,y_train4,y_test4= train_test_split(X_scale4,y4, random_state=42)
Model4.fit(X_train4,y_train4)
X_train_pred4=Model4.predict(X_train4)
X_test_pred4=Model4.predict(X_test4)
print("This is training accuracy :", accuracy_score(X_train_pred4, y_train4))
print("This is testing accuracy :", accuracy_score(X_test_pred4, y_test4))
```
